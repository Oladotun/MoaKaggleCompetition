# MOA LASSO AND PCA ANALYSIS

#### For the mechanism of analysis for a new drug we intend to perform a feature selection to know what features are important for analysis

#### The output consist of about 403 columns that needs to be reduced using principal component analysis and from there we can perform feature selection on the input variables

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import Lasso
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
%matplotlib inline
```

```python
df = pd.read_csv("~/Desktop/MoaKaggleCompetition/DataSets/train_features.csv")
target = pd.read_csv("~/Desktop/MoaKaggleCompetition/DataSets/train_targets_scored.csv")
```

```python
df.describe()
```

#### PCA on target variables

```python
pca = PCA(n_components=1)
principalComponents = pca.fit_transform(target.drop("sig_id", axis=1))
targetDf = pd.DataFrame(data = principalComponents
             , columns = ['target'])
```

```python
targetDf.head()
```

```python
targetDf = targetDf["target"].apply(lambda x: 0 if x < 0 else 1)
```

```python
targetDf.head()
```

```python
len(targetDf[targetDf==1])
```

```python
df.head()
```

```python
df_without_id = df.drop("sig_id", axis=1)
```

```python
dummy = pd.get_dummies(df_without_id [["cp_type","cp_dose"]])
```

```python
df_without_id = df_without_id.merge(dummy, left_index=True, right_index = True)
```

```python
df_without_id = df_without_id.drop(columns=["cp_type","cp_dose"])
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(
    penalty='l1',
    solver='saga')

model.fit(df_without_id, targetDf.values)
```

```python
res =pd.DataFrame(zip(df_without_id.columns, np.transpose(model.coef_[0])), columns=['features', 'coef'])
```

```python
fil_res = res[res["coef"] != 0]
```

```python
len(fil_res)
```

```python
fil_res.sort_values(by=['coef'], ascending=False, inplace= True)
```
